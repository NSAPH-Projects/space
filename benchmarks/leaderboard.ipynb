```python
import jsonlines
import glob
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import itertools

# show all rows and columns of dataframes
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Load Benchmark Results

```python
algomap = {
    "spatial": "\\textsc{spatial}",
    "spatialplus": "\\textsc{spatial+}",
    "gcn": "\\textsc{gcnn}",
    "dapsm": "\\textsc{dapsm}",
    "ols": "\\textsc{ols}",
    "gmerror": "\\textsc{gmerror}",
    "s2sls-lag1": "\\textsc{s2sls-lag1}",
    "s2sls-lag2": "\\textsc{s2sls-lag2}",
    "xgboost": "\\textsc{xgboost}",
}

directory_path = "benchmarks/results"
file_list = glob.glob("**/*.jsonl", root_dir=directory_path, recursive=True)
print(f"Found {len(file_list)} files")

results = []
for file in file_list:
    with jsonlines.open(f"{directory_path}/{file}", "r") as reader:
        results.extend(list(reader))

results = pd.DataFrame(results)

results = (
    results
    .assign(confounding_erf=[x["erf"] for x in results["confounding"]])
    .assign(confounding_ate=[x["ate"] for x in results["confounding"]])
    .assign(confounding_ite=[x["ite"] for x in results["confounding"]])
    .assign(confounding_importance=[x["importance"] for x in results["confounding"]])
    # .query("algo != 's2sls-lag2' and algo != 'xgboost'")
    .assign(algo=results.algo.map(algomap))
)
smooth_thrsh= results.smoothness.median()
conf_erf_thrsh = np.nanquantile(results.confounding_erf, 0.5)
conf_ate_thrsh = np.nanquantile(results.confounding_ate, 0.5)
conf_ite_thrsh = np.nanquantile(results.confounding_ite, 0.5)
conf_imp_thrsh = np.nanquantile(results.confounding_importance, 0.5) 
results = (
    results
    .assign(smooth_level=["high" if x > smooth_thrsh else "low" for x in results["smoothness"]])
    .assign(conf_ate_level=["high" if x > conf_ate_thrsh else "low" for x in results["confounding_ate"]])
    .assign(conf_erf_level=["high" if x > conf_erf_thrsh else "low" for x in results["confounding_erf"]])
    .assign(conf_ite_level=["high" if x > conf_ite_thrsh else "low" for x in results["confounding_ite"]])
    .assign(conf_imp_level=["high" if x > conf_imp_thrsh else "low" for x in results["confounding_importance"]])
)
results.head()
```

```python
# visualize smoothness and confounding distrib
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(1, 5, figsize=(11, 2))
sns.histplot(results["smoothness"], bins=20, ax=ax[0], kde=True); ax[0].set_xlabel("Smoothness");
ax[0].axvline(smooth_thrsh, color="red")
sns.histplot(results["confounding_ate"], ax=ax[1], bins=20, kde=True); ax[1].set_xlabel("ATE")
ax[1].axvline(conf_ate_thrsh, color="red")
sns.histplot(results["confounding_erf"], ax=ax[2], bins=20, kde=True); ax[2].set_xlabel("ERF")
ax[2].axvline(conf_erf_thrsh, color="red")
sns.histplot(results["confounding_ite"], ax=ax[3], bins=20, kde=True); ax[3].set_xlabel("ITE")
ax[3].axvline(conf_ite_thrsh, color="red")
sns.histplot(results["confounding_importance"], ax=ax[4], bins=20, kde=True); ax[4].set_xlabel("FeatImp")
ax[4].axvline(conf_imp_thrsh, color="red")
plt.suptitle("Distribution of smoothness and confounding scores")
plt.savefig("Distribution of smoothness and confounding scores.")
```

## Paper Table

```python
def bold_best(x):
    cols = x.columns[-2:]
    for col in cols:
        v = x[col]
        i = np.argmin(v.values)
        x[col].iloc[i] = "\\bf " + v.iloc[i]
    return x


def normalize_count(x):
    x["N"] = int(np.median(x["N"].values))
    return x


def pm(x):
    m = np.nanmean(x)
    n = np.sum(~np.isnan(x))
    ci = np.nanstd(x) / np.sqrt(n) * 1.96
    return f"{m:.2f}  " + "{\\small " + f"{ci:.2f}" + "}"


binary_total = (
    results[results.binary]
    .groupby(["smooth_level", "conf_erf_level", "algo"])
    .agg( bin_ate=("ate", pm), bin_ite=("ite", pm))
    .fillna("n/a")
    .rename_axis(index=["Smoothness", "Confounding", "Method"])
    .reset_index()
    .groupby(["Smoothness", "Confounding"], group_keys=False)
    .apply(bold_best)
    # .groupby(["Smoothness", "Confounding"], group_keys=False)
    # .apply(normalize_count)
    .set_index(["Smoothness", "Confounding", "Method"])
)

cont_total = (
    results[~results.binary]
    .groupby(["smooth_level", "conf_erf_level", "algo"])
    .agg(cont_erf=("erf", pm), cont_ite=("ite", pm))
    .fillna("n/a")
    .rename_axis(index=["Smoothness", "Confounding", "Method"])
    .reset_index()
    .groupby(["Smoothness", "Confounding"], group_keys=False)
    .apply(bold_best)
    # .groupby(["Smoothness", "Confounding"], group_keys=False)
    # .apply(normalize_count)
    .set_index(["Smoothness", "Confounding","Method"])
)

# # outer join
fin = pd.merge(binary_total, cont_total, how="outer", left_index=True, right_index=True)
fin.fillna("n/a", inplace=True)
fin.columns = pd.MultiIndex.from_tuples(
    [
        ("\\textsc{binary}", "\\textsc{ate}"),
        ("\\textsc{binary}", "\\textsc{ite}"),
        ("\\textsc{continuous}", "\\textsc{erf}"),
        ("\\textsc{continuous}", "\\textsc{ite}"),
    ]
)
fin
```

```python
os.makedirs("benchmarks/tex", exist_ok=True)
fin.to_latex("benchmarks/tex/leaderboard.tex", index=True, escape=False)

# prepend begin document use package booktabs etc to the above latex
with open("benchmarks/tex/template.tex", "r") as f:
    template = f.read()

with open("benchmarks/tex/leaderboard.tex", "r") as f:
    table = f.read()

with open("benchmarks/tex/leaderboard.tex", "w") as f:
    f.write(template.replace("<table>", table))

# compile latex to pdf
import subprocess
subprocess.run(
    [
        "latexmk",
        "-pdf",
        "-jobname=benchmarks/tex/leaderboard",
        "benchmarks/tex/leaderboard.tex",
    ]
)

# remove .aux, .fdb_latexmk, .fls, .log files from benchmarks/tex
for ext in ["aux", "fdb_latexmk", "fls", "log", "gz"]:
    for f in glob.glob(f"benchmarks/tex/*.{ext}"):
        os.remove(f)
```

## Visual Comparisons per Confounding Score and Smoothness Level

```python
#white grid and pastel palette
sns.set_theme(style="whitegrid", palette="pastel")

plot_data = (
    results
    .assign(algo=lambda x: x.algo.str.replace("\\textsc{", "").str.replace("}", ""))
    .melt(
        id_vars=["smooth_level", "conf_erf_level", "env", "algo", "binary"],
        value_vars=["erf", "ite"],
        var_name="estimand",
        value_name="Causal estimate error",
    )
    .assign(estimand=lambda x: x.estimand.str.upper())
    # map binary to binary and cont
    .assign(binary=lambda x: x.binary.map({True: "binary", False: "continuous"}))
    # captitalize estimand
    .assign(estimand=lambda x: x.estimand.str.upper())
    # make all combinations of binary and estimand
    .assign(task=lambda x: x.binary + " | " + x.estimand)
    # make all combs of smooth and conf
    .assign(challenge=lambda x: [f"{a} smooth. | {b} conf." for a, b in zip(x.smooth_level, x.conf_erf_level)])
)
algo_names = sorted(plot_data.algo.unique())

# margin titles on left
palette = dict(zip(algo_names, sns.color_palette("tab10", n_colors=len(algo_names))))

g = sns.FacetGrid(
    plot_data,
    col="task",
    row="challenge",
    sharex=False,
    sharey=False,
    height=3.2,
    aspect=1.0,
    # hue="algo",
)
sns.set(font_scale=0.95)

# def fixed_barplot(x, y, hue, label=None, **kwargs):
#     sns.barplot(x=x, y=y, hue=hue, labels=[label], **kwargs)


# # make a barplot with the ci as error bar.
g.map_dataframe(
    sns.barplot,
    "algo",
    y="Causal estimate error",
    capsize=.2,
    palette=palette,
    errwidth=1.5,
    width=1.0,
    dodge=False,
    order=algo_names,
)

# Now modify the titles
for ax in g.axes.flat:
    title = ax.get_title()
    new_title = title.replace("challenge = ", " ").replace("task = ", "")
    ax.set_title(new_title)
    # make x axis font smaller
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=12)
    # make y ticks smaller
    ax.set_xlabel("")
    ax.tick_params(axis='y', labelsize='small')

# legend = g.add_legend(bbox_to_anchor=(0.5, 1.05), ncol=8, fontsize=10)
g.fig.subplots_adjust(hspace=0.5, wspace=0.25)
plt.savefig("benchmarks/figures/results_by_task.pdf", bbox_inches="tight")
```

## Visualization of Pairwise Comparisons

```python
analysis_data = (
    results
    .assign(algo=lambda x: x.algo.str.replace("\\textsc{", "").str.replace("}", ""))
)
analysis_data.head()
```

```python
pairs_data = []

for (algo1, algo2) in itertools.combinations(algo_names, 2):
    subset1 = analysis_data.query(f"algo == '{algo1}'")
    subset2 = analysis_data.query(f"algo == '{algo2}'")
    merged = (
        pd.merge(subset1, subset2, on=['env', 'dataset_id'], how="outer")
        .assign(Comparison=f"{algo1} - {algo2}")
        .assign(ERF=lambda x: x.erf_x - x.erf_y)
        .assign(ITE=lambda x: x.ite_x - x.ite_y)
        .melt(
            id_vars=["env", "Comparison", "dataset_id"],
            value_vars=["ERF", "ITE"],
            var_name="Estimand",
            value_name="Difference",
        )
    )
    pairs_data.append(merged)
pairs_data = pd.concat(pairs_data)
pairs_data.head()

# Create FacetGrid
sns.set_theme(style="whitegrid")
g = sns.FacetGrid(
    pairs_data[pairs_data.Estimand == "ERF"],
    col="Comparison",
    col_wrap=6,
    sharex=False,
    sharey=False,
    height=3,
)
g.map_dataframe(sns.kdeplot, x="Difference")


# Add vertical lines and ribbons for mean and confidence interval
def add_mean_and_ci(*args, **kwargs):
    data = kwargs.pop("data")
    ax = plt.gca()
    x = data.Difference.values
    mean = np.nanmean(x)
    ci = np.nanstd(x) / np.sqrt(len(x[~np.isnan(x)]))

    ax.axvline(mean, color="red", linestyle="-", label="Mean")
    ax.fill_betweenx(
        [0, ax.get_ylim()[1]],
        mean - ci,
        mean + ci,
        alpha=0.3,
        color="red",
        label="95% CI",
    )
    ax.axvline(0.0, color="black", linestyle="--", label="Zero")


g.map_dataframe(add_mean_and_ci, x="Difference")

for ax in g.axes.flat:
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_title(ax.get_title().replace("Comparison = ", ""))

# Create a global legend
handles, labels = ax.get_legend_handles_labels()
g.fig.legend(handles, labels, loc="upper center", ncol=3, fontsize=12)

# Adjust subplot to fit global legend on top
g.fig.subplots_adjust(top=0.95, wspace=0.1)
```

## Formal Statistical Analysis

An installation of R and the packages `multcomp` and `lme4` is needed for the following formal test.

```python
%load_ext rpy2.ipython
```

```python
%%R -i analysis_data
library(lme4)
library(multcomp)

df = analysis_data
df$algo <- factor(df$algo, levels=c("ols", "spatial", "spatial+", "gcnn", "dapsm", "gmerror", "s2sls-lag1", "s2sls-lag2", "xgboost"))

model <- lmer(erf ~ algo + (1|env:dataset_id), data = df)
print(summary(glht(model, linfct = mcp(algo = "Tukey"))))
```

```python
%%R 
model <- lmer(ite ~  + algo + (1|env:dataset_id), data = df)
print(summary(model))
print(summary(glht(model, linfct = mcp(algo = "Tukey"))))
```

```python
%%R 
model <- lm(erf ~ confounding_erf, data = df)
print(summary(model))

model <- lm(ite ~ confounding_ite, data = df)
print(summary(model))

model <- lm(erf ~ confounding_importance, data = df)
print(summary(model))

model <- lm(ite ~ confounding_importance, data = df)
print(summary(model))
```

```python
%%R 
model <- lmer(erf ~ algo + algo * smoothness  + (1 | env), data = df)
print(summary(model))

model <- lmer(ite ~ algo + algo * smoothness  + (1 | env), data = df)
print(summary(model))
```
